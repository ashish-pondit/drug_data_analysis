```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
drug=pd.read_csv('drug_user.csv')
```

```python
drug['user_drug']=drug.user_drug.str.lower()
```

```python
df=drug.user_drug.value_counts()
```

```python
df=df.head()
```

```python
df=df.sort_values()
```

# Analysing Top most 5 drugs 

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df.index,y=df.values)
plt.title('Most used top 5 drugs')
plt.xlabel('Drug names')
plt.ylabel('No of users')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.pie(df.values,labels=df.index,
        autopct='%1.1f%%',
        explode=[0.01,0.01,0.01,0.01,0.01],
        pctdistance=0.5,
        colors=['#FE666F','#FE66D7','#AB66FE','#6689FE','#66FEB0'],
        startangle=90)
plt.legend(labels=df.index)
plt.title('Most used top 5 medichine')
plt.show()
```
